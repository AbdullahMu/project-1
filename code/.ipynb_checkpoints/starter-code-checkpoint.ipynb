#### 7. Rename columns
Change the names of the columns to more expressive names so that you can tell the difference the Driving Licenses  columns and the & Traffic Accidents columns. Your solution should map all column names being changed at once (no repeated singular name-changes). **We will be combining these two datasets, and so you should name columns in an appropriate way**.

**Guidelines**:
- Column names should be all lowercase (you will thank yourself when you start pushing data to SQL later in the course)
- Column names should not contain spaces (underscores will suffice--this allows for using the `df.column_name` method to access columns in addition to `df['column_name']`.
- Column names should be unique and informative (the only feature that we actually share between dataframes is the state).

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def standard_deviation(x):
    return np.sqrt(np.mean(abs(x - np.mean(x))**2))
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'Driving_Licenses_x': 120, 'Traffic_Accidents_x': 120, ...}`

````python
cols = merged_dataframe.columns.tolist()
standard_deviation_dict = {i: standard_deviation(merged_dataframe[i]) for i in cols if str(merged_dataframe[i].dtypes) != 'object'}
standard_deviation_dict
````

Do your manually calculated standard deviations match up with the output from pandas `describe`? What about numpy's `std` method?

````python
for i in cols:
    if str(merged_dataframe[i].dtypes) != 'object':
        print('standard deviations of', i, 'match?', float(description[['std']].loc[i]) == standard_deviation_dict[i])  
````

**Answer:** No, they don't match because NumPy uses N degrees of freedom, while pandas uses Bessel's correction (N-1 egrees of freedom)

````python
licenses_and_mean_df.rename(index=str, 
                            columns={"driving_licenses_x": "no._driving_licenses", 
                                     "driving_licenses_y": "driving_licenses_annual_mean"}, inplace=True)

licenses_and_mean_df.drop(columns=['longitude', 'latitude'], inplace=True)

licenses_and_mean_df['driving_licenses_annual_mean'] = licenses_and_mean_df['driving_licenses_annual_mean'].map(lambda x: round(x,2))

licenses_and_mean_df.head()
````

````python
licenses_more_than_mean = licenses_and_mean_df[licenses_and_mean_df['no._driving_licenses']>=licenses_and_mean_df['driving_licenses_annual_mean']]
licenses_more_than_mean.head()
````

````python
grouped_licenses = licenses_more_than_mean.groupby(['year', 'region'])
grouped_licenses_df = grouped_licenses.first()
grouped_licenses_df
````

````python
grouped_licenses_df.reset_index()['region'].value_counts()
````

````python
###ANSWER###
#What are the regions that have more Driving Licenses issued in each year than the year average?

THE_REGION = grouped_licenses_df.reset_index()['region'].value_counts().idxmax()
THE_YEARS = grouped_licenses_df.reset_index()['year'][grouped_licenses_df.reset_index()['region'] == THE_REGION].tolist()

print('the only region whose issued number of Driving Licenses is more than annual average each year is', THE_REGION, '\n')

print(THE_REGION, 'has surpassed the annual average of licensce issuance in the following years:')

print(*THE_YEARS, sep=", ")
````

````python
#second region
THE_REGION_S = ['Riyadh', 'Eastern Region' ]

print('the next regions with highest number of Driving Licenses issuance more than annual average are', " and ".join(THE_REGION_S), '\n')


THE_YEARS1 = grouped_licenses_df.reset_index()['year'][grouped_licenses_df.reset_index()['region'] == THE_REGION_S[0]].tolist()

THE_YEARS2 = grouped_licenses_df.reset_index()['year'][grouped_licenses_df.reset_index()['region'] == THE_REGION_S[1]].tolist()

print(THE_REGION_S[0], 'has missed the mark on years', THE_YEARS[THE_YEARS not in THE_YEARS1])
print(THE_REGION_S[1], 'has missed the mark on years', THE_YEARS[THE_YEARS not in THE_YEARS2])
````

````python
def subplot_scatterplots(dataframe, list_of_x, list_of_y, year=None):
    
    nrows = 1 if len(list_of_x) == 1 else int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    ncols = 1 if len(list_of_y) == 1 else 2
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(8,6) ) # You'll want to specify your figsize
    
    if nrows != 1:
        ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
        for i, x_y in enumerate(zip(list_of_x, list_of_y)): # Gives us an index value to get into all our lists
            if year:
                ax[i].scatter(dataframe[dataframe['year'] == int(year)][x_y[0]], dataframe[dataframe['year'] == int(year)][x_y[1]]) 
                ax[i].set_xlabel(x_y[0])
                ax[i].set_ylabel(x_y[1])
                ax[i].set_title(x_y[1] + ' vs. ' + x_y[0] + ' for ' + str(year))
            else:
                ax[i].scatter(dataframe[x_y[0]], dataframe[x_y[1]]) 
                ax[i].set_xlabel(x_y[0])
                ax[i].set_ylabel(x_y[1])
                ax[i].set_title(x_y[1] + ' vs. ' + x_y[0])   
    else:
    
        if year:
            ax.scatter(dataframe[dataframe['year'] == int(year)][list_of_x], dataframe[dataframe['year'] == int(year)][list_of_y]) 
            ax.set_xlabel(str(list_of_x).strip('[]\''))
            ax.set_ylabel(str(list_of_y).strip('[]\''))
            ax.set_title(str(list_of_y).strip('[]\'') + ' vs. ' + str(list_of_x).strip('[]\'') + ' for ' + str(year))
        else:
            ax.scatter(dataframe[list_of_x], dataframe[list_of_y]) 
            ax.set_xlabel(str(list_of_x).strip('[]\''))
            ax.set_ylabel(str(list_of_y).strip('[]\''))
            ax.set_title(str(list_of_y).strip('[]\'') + ' vs. ' + str(list_of_x).strip('[]\''))   
````

````python
#let's hypothesize that the number of accidents is not correlated with driver licenses issuance. 
#We can argue that most accident are not caused by newly licensed driver but rather ALL drivers.
#But let's test our hypothesis statistically

# Ho: number of accidents is correlated with driver licenses issuance
# H1: number of accidents is not correlated with driver licenses issuance

#since correlation takes values within the range [-1, 1], then no need for scaling
#set threshold = 0.1 for rejection area

merged_dataframe[['driving_licenses', 'No. of Accidents']].corr()
````

````python
# Ho: number of accidents is correlated with driver licenses issuancenumber of accidents is not correlated with driver licenses issuance (correlation < threshold = 0.1)
# H1: number of accidents is not correlated with driver licenses issuance (correlation >= threshold = 0.1)

# statistically speaking, the numbers are strongly correlated, our hypothesis is wrong, 
# the reason is that both of number of accidents and driver licenses issuance must be driven by population growth.
````
